```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%load_ext autoreload
%autoreload 2
%matplotlib inline

from pymongo import MongoClient
import pymongo
client = MongoClient()
db = client.codemd

from scipy.interpolate import spline
```

```python
from metrics_builder import MetricsBuilder
```

```python
rails = db['rails']
```

```python
builder = MetricsBuilder(rails)
rails_info = builder.defect_rates()
```

```python
rails_info
```

```python
df = pd.DataFrame(rails_info)
```

```python
bug_info = df['bug_count']
```

```python
sorted_bugs = bug_info.sort_values(ascending=False)
```

```python
xnew = np.linspace(0,12,300)

power_smooth = spline(sorted_bugs,2,xnew)

plt.plot(xnew,power_smooth)
plt.show()
```
